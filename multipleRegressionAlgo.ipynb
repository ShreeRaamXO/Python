```python
df = pd.read_csv('bRoomAgeDs.csv')
df
```

```python
# since one value is null , we replace it with mean / median / mode
BroomMed = df.bRooms.median()
BroomMed
```

```python
df.bRooms = df.bRooms.fillna(BroomMed)
df
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(df.drop('Price', axis='columns'),df.Price)
```

```python
reg.predict([[2500 , 3 , 40]])
```
